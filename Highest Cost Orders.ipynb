Find the customer with the highest total order cost between 2019-02-01 to 2019-05-01. Output their first name, total cost of their items, and the date.

For simplicity, you can assume that every first name in the dataset is unique.

first_nameobject

phone_numberobject
orders

cust_idint64
order_datedatetime64[ns]

```python
# Import your libraries
import pandas as pd

# Start writing code
customers.head()
df = pd.merge(customers,orders,how='left',left_on='id',right_on='cust_id')
df['order_date'] = pd.to_datetime(df['order_date'])
df['total']= df['order_cost']*df['order_quantity']
merge = df[df['order_date'].between('2019-02-1', '2019-05-1')]
result = merge.groupby(['first_name', 'order_date'])['total'].sum().to_frame('max_cost').reset_index() 
result = result[result.max_cost == result.max_cost.max()] 
result['order_date'] = pd.to_datetime(result['order_date'],format='%Y-%m-%d').dt.strftime('%Y-%m-%d')
result
```
